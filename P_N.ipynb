```python
#@markdown START!

from PIL import Image
import requests
import io
import random
from collections import OrderedDict

# 이미지와 영단어 매칭 퀴즈 데이터
quiz_data = OrderedDict([
    ("loyalty", "https://raw.githubusercontent.com/HonestHabin/DL23/main/loyalty.jpg"),
    ("chemical", "https://raw.githubusercontent.com/HonestHabin/DL23/main/chemical.jpg"),
    ("conservation", "https://raw.githubusercontent.com/HonestHabin/DL23/main/conservation.jpg"),
    ("corrosion", "https://raw.githubusercontent.com/HonestHabin/DL23/main/corrosion.jpg"),
    ("damage", "https://raw.githubusercontent.com/HonestHabin/DL23/main/damage.jpg"),
    ("generation", "https://raw.githubusercontent.com/HonestHabin/DL23/main/generation.jpg"),
    ("x-ray photography", "https://raw.githubusercontent.com/HonestHabin/DL23/main/x-ray%20photography.jpg"),
    ("cultural heritage", "https://raw.githubusercontent.com/HonestHabin/DL23/main/cultural%20heritage.jpg"),
    ("glue", "https://raw.githubusercontent.com/HonestHabin/DL23/main/glue.jpeg"),
    ("document", "https://raw.githubusercontent.com/HonestHabin/DL23/main/document.jpeg"),
    ("pagoda", "https://raw.githubusercontent.com/HonestHabin/DL23/main/pagoda.jpeg"),
    ("reward", "https://raw.githubusercontent.com/HonestHabin/DL23/main/reward.jpeg"),
    ("introduction", "https://raw.githubusercontent.com/HonestHabin/DL23/main/introduction.jpg")
])

# 플레이어 점수
score = 0

# 퀴즈 게임 함수
def play_quiz():
    global score  # 전역 변수로 점수에 접근하기 위해 global 키워드 사용

    # 퀴즈 문제 생성
    word = next(iter(quiz_data))
    image_url = quiz_data[word]

    # 이미지 다운로드
    response = requests.get(image_url)
    image = Image.open(io.BytesIO(response.content))

    # 이미지 표시
    image.show()

    # 이미지 크기 조정
    fixed_size = (200, 200)  # 원하는 크기로 수정 가능
    image = image.resize(fixed_size)

    # 사용자 입력 받기
    answer = input("영단어를 입력하세요: ")

    # 정답 확인
    if answer.lower() == word:
        score += 1  # 정답을 맞춘 경우 점수 1 증가
        print("정답입니다!")
    else:
        print("오답입니다! 정답은", word, "입니다.")

    # 다음 문제를 위해 퀴즈 데이터에서 현재 문제 제거
    quiz_data.pop(word)

# 퀴즈 플레이어 실행
while quiz_data:
    play_quiz()
    
    # 퀴즈 데이터가 모두 소진되었을 경우 반복 종료
    if not quiz_data:
        break

        # 플레이어 점수 출력
print("최종 점수:", score)

# 퀴즈 종료
print("퀴즈 종료!")
```

```python
#@markdown
import ipywidgets as widgets
from IPython.display import display

Question = "1" #@param ["1", "2", "3", "4", "5", "6"]

# 문제와 답
questions_and_answers = {
    1: ("""Conservation science plays a role in revealing historical truth and hidden stories of cultural heritage.""", True),
    2: ("""Baekje geumdonggwanmo caps were often given to local leaders as rewards for their loyalty.""", True),
    3: ("""X-ray photography was used by conservation scientists to determine the original form of the cap.""", True),
    4: ("""During the conservation process, foreign matter and rust were added to the surface of the cap.""", False),
    5: ("""The cap was repaired and any damaged parts were replaced during the restoration process.""", True),
    6: ("""The cap lost its original form and shine after the conservation process.""", False),
}
# False 문제의 True 답
True_of_False_questions = {
    4: ("""During the conservation process, conservation scientists removed foreign matter and rust from the surface and controlled corrosion with a chemical treatment."""),
    6: ("""At last, the cap regained its original form and shine.""")
}
# 문제 선택
if Question.isdigit() and int(Question) in questions_and_answers:
    question_number = int(Question)
    question_text = questions_and_answers[question_number][0]
    print(question_text)

    # 정답 입력
    while True:
        answer_input = input("Enter your answer (True/T or False/F): ")
        
        if answer_input in ["True", "true", "T", "t", "False", "false", "F", "f"]:
            break
        else:
            print("Invalid answer. Please enter True/T or False/F.")
    
    # 문제 풀이
    solve_question(question_number, answer_input)



def solve_question(question, answer_input):
    question_text, answer = questions_and_answers[question]
    

    # 정답 확인
    if answer_input in ["True", "true", "T", "t"]:
        user_answer = True
    elif answer_input in ["False", "false", "F", "f"]:
        user_answer = False
    else:
        print("Invalid answer.")
        return
    
    # 정답과 오답 텍스트 출력
    if user_answer == answer:
        print("Correct!")
    else:
        print("Wrong answer.")
        if answer is False:
            if question in True_of_False_questions:
                print(True_of_False_questions[question])
```

```python
#@markdown 금동관모 복원 순서 퀴즈
import random

# 주어진 순서대로 단계를 나열한 리스트
steps = [
    "Conservation scientists used X-ray photography to determine the original form.",
    "The cap was broken down into parts.",
    "Conservation scientists removed foreign matter and rust from the surface.",
    "Conservation scientists controlled corrosion with a chemical treatment.",
    "All the parts were glued together and any damaged parts were repaired.",
    "The cap regained its original form and shine."
]

# 답을 섞어줌
random.shuffle(steps)

# 순서 맞추기 퀴즈 문제 출력
print("다음은 금동관모 복원 과정의 단계들입니다. 올바른 순서대로 나열해 보세요.\n")

for i, step in enumerate(steps):
    print(f"{i+1}. {step}")

print("\n순서를 선택하세요. (숫자로 입력해 주세요)")

# 올바른 순서
correct_order = [
    "Conservation scientists used X-ray photography to determine the original form.",
    "The cap was broken down into parts.",
    "Conservation scientists removed foreign matter and rust from the surface.",
    "Conservation scientists controlled corrosion with a chemical treatment.",
    "All the parts were glued together and any damaged parts were repaired.",
    "The cap regained its original form and shine."
]

# 사용자로부터 순서 입력 받기
user_order = []
for i in range(len(steps)):
    while True:
        try:
            step_number = int(input(f"{i+1}. 단계 번호: "))
            if step_number < 1 or step_number > len(steps) or step_number in user_order:
                raise ValueError
            break
        except ValueError:
            print("유효한 단계 번호를 입력하세요.")

    user_order.append(steps[step_number-1])

# 결과 출력
print("\n---결과---\n")
if user_order == correct_order:
    print("축하합니다! 올바른 순서로 단계를 나열하셨습니다.")
else:
    print("순서가 잘못되었습니다. 올바른 순서는 다음과 같습니다:")
    for correct_step in correct_order:
        print("- ", correct_step)
```

✅ Read the next three questions and discuss them with the group. When the discussion is over, write down your answers.

◾ Why is it important to preserve cultural heritage?

◾ How does cultural heritage contribute to our understanding of history and sense of identity?:

◾ Can you think of other examples where conservation science has been successful in preserving cultural heritage?

```python
#@markdown
class Debate:
    def __init__(self, questions):
        self.questions = questions
        self.answers = []
    
    def add_answer(self, answer):
        self.answers.append(answer)
    
    def print_answers(self):
        for i, question in enumerate(self.questions):
            print(f"Question {i+1}: {question}")
            print(f"Answer: {self.answers[i]}")
            print()
    
    def ask_questions(self):
        for question in self.questions:
            answer = input(f"{question}: ")
            self.add_answer(answer)


questions = [
    "Why is it important to preserve cultural heritage?",
    "How does cultural heritage contribute to our understanding of history and sense of identity?",
    "Can you think of other examples where conservation science has been successful in preserving cultural heritage?"
]
debate = Debate(questions)

print("please answer the questions.")

debate.ask_questions()

print("                              ")

debate.print_answers()  
```
